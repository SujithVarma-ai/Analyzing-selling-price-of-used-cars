Drop Column

```python
dataset = dataset.drop(columns=['name'])
```

```python
dataset.head()
```

Add column

```python
dataset['Current Year'] = 2026
```

```python
dataset.head()
```

How to calculate the car that how many years old it is

```python
dataset['Years Old'] = dataset['Current Year'] - dataset['year']
```

```python
dataset.head()
```

```python
dataset = dataset.drop(columns=['Current Year', 'year'], errors='ignore')
```

```python
dataset.head()
```

Encoding Categorical variables

```python
dataset.select_dtypes(include='object').columns
```

One hot encoding

```python
dataset = pd.get_dummies(data = dataset, drop_first=True)
```

```python
dataset.head()
```

matrix of feature

```python
x = dataset.drop(columns='selling_price')
y = dataset['selling_price']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

Building the Model

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor()
regressor.fit(x_train, y_train)
```

```python
y_pred = regressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

Hyperparameter tuning

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
parameter = {
    'n_estimators': np.arange(100, 1200, 100),
    'criterion': ['squared_error', 'absolute_error'],
    'max_depth': [5, 10, 20,40],
    'min_samples_split': [2, 5, 10, 15, 100],
    'min_samples_leaf': [1, 2, 5, 10],
    'max_features': ['sqrt', 'log2']
}
```

```python
parameter
```

```python
random_cv = RandomizedSearchCV(estimator=regressor, param_distributions=parameter, scoring='neg_mean_squared_error', n_iter=10,
                               cv=5, verbose=2, random_state=0)
```

```python
random_cv.fit(x_train, y_train)
```

```python
random_cv.best_params_
```
